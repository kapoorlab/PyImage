```python
import sys
sys.path.insert(0, "/Users/aimachine/PyImage/utils")
import numpy as np
from scipy import signal
from scipy.fftpack import fftfreq
import matplotlib.pyplot as plt
import os
from scipy.signal import blackman
from tifffile import imread
from scipy.fftpack import fft
from matplotlib.colors import LogNorm
from Normalize import normalizeFloat, normalizeMinMax, Path, save_tiff_imagej_compatible
```

```python
targetdir = '/Users/aimachine/Documents/JLMCurvature/FFTImageTest/'
sourceimage = '/Users/aimachine/Documents/JLMCurvature/FFTImageTest/FullResolution.tif'
y = imread(sourceimage)
base = os.path.split(sourceimage)[-1]

Filename = os.path.splitext(base)
```

```python
#Enter time unit in seconds
Time_unit = 5
#Enter start y pixel unit for the section
StartSection = 0
#Enter end y pixel unit for section
EndSection = 100
#Range of Plot in Hertz
maxXrange = 0.05
```

```python
plt.figure()
plt.imshow(y, plt.cm.Spectral)
plt.title('Original image')
Xrange = y.shape[0]
Yrange = y.shape[1]
print(Xrange, " ", Yrange)
width = int(EndSection - StartSection)
block = int(Xrange / width)

Start = StartSection
End = EndSection

axes = 'XY'
```

```python
#FFT without filter

def doFFT(image, block):
   addedfft = 0 
   
   for i in range(image.shape[0]):
    
      strip = image[i,:]
      fftresult = fft(strip)
      addedfft += np.abs(fftresult)  
   addedfft/=image.shape[0]
   plt.figure()
   PointsSample = image.shape[1] 
   xf = fftfreq(PointsSample, Time_unit)
   
   plt.semilogy(xf[1:int(PointsSample//2)], 2.0/PointsSample * (addedfft[1:int(PointsSample//2)]), '-b')
   plt.xlim(xmax = maxXrange)
   plt.grid()
   plt.legend(['FFT (hertz) Peak(seconds) = 1 / Peak(hertz)']) 
   plt.savefig(targetdir + Filename[0] + str(block) + '.png')
   return addedfft  
```

```python
JoinFFt = np.zeros([block, Yrange])
for i in range(block):
 plt.figure()
 smally = y[StartSection:EndSection, :]
 plt.imshow(smally, plt.cm.Spectral)
 plt.title('Image Segment of interest')

 StartSection = EndSection
 EndSection += width    
 onedimg = doFFT(smally, i)
 JoinFFt[i,:] = onedimg
StartSection = Start
EndSection = End
plt.figure()
plt.imshow(JoinFFt, plt.cm.Spectral)
save_tiff_imagej_compatible((targetdir + 'FFTResult' + Filename[0] + ".tif")  , JoinFFt, axes)
```

```python
#FFT with filter
def doFilterFFT(image, block):
   addedfft = 0 
   PointsSample = image.shape[1] 
   for i in range(image.shape[0]):
      w = blackman(PointsSample)
      strip = image[i,:]
      fftresult = fft(w * strip)
      addedfft += np.abs(fftresult)  
   addedfft/=image.shape[0]
   plt.figure()
   
   xf = fftfreq(PointsSample, Time_unit)
   
   plt.semilogy(xf[1:int(PointsSample//2)], 2.0/PointsSample * (addedfft[1:int(PointsSample//2)]), '-r')
   plt.xlim(xmax = maxXrange)
   plt.grid()
   plt.legend(['FFT (hertz) Peak(seconds) = 1 / Peak(hertz)']) 
   plt.savefig(targetdir + 'Filtered' + Filename[0] + str(block) + '.png')
   return addedfft 
```

```python


JoinFFt = np.zeros([block, Yrange])
for i in range(block):
 plt.figure()
 fsmally = y[StartSection:EndSection, :]
 plt.imshow(fsmally, plt.cm.Spectral)
 plt.title('Image Segment of interest')
    
 StartSection = EndSection
 EndSection += width    
 onedimg = doFilterFFT(fsmally, i)
 JoinFFt[i,:] = onedimg
StartSection = Start
EndSection = End
plt.figure()
plt.imshow(JoinFFt, plt.cm.Spectral)
save_tiff_imagej_compatible((targetdir + 'FilteredFFTResult' + Filename[0] + ".tif") , JoinFFt, axes)
```

```python
import sys
sys.path.insert(0, "/Users/aimachine/PyImage/utils")
import numpy as np

import importlib
import matplotlib.pyplot as plt
import os
import glob
import skimage
from skimage import measure
from skimage import filters

from skimage.filters import threshold_otsu
from tifffile import imread
import scipy
from scipy import ndimage
from skimage.measure import compare_ssim as ssim
```

```python
basedir = '/Users/aimachine/Downloads/GrantApplication/StructureSimilarity/'
FileLow = basedir + '40x_bin1_05_5min_s1-4Cg_s5-6esg_1_w2CSU-488_s1_t125LOW.tif'
FileGT =  basedir + '40x_bin1_05_5min_s1-4Cg_s5-6esg_1_w2CSU-488_s1_t125GT.tif'
FileRestored = basedir + 'BorialisS1S2FlorisMidNoiseModel_40x_bin1_05_5min_s1-4Cg_s5-6esg_1_w2CSU-488_s1_t125.tif'
```

```python
Low = imread(FileLow)
GT = imread(FileGT)
Restored = imread(FileRestored)
Time = Low.shape[0]
```

```python
def mse(x, y):
    return np.linalg.norm(x - y)
```

```python
LowMSE = []
GTMSE = []
RestoredMSE = []

LowSSIM = []
GTSSIM = []
RestoredSSIM = []

for j in range(Time):
      smallLowimg = Low[j,:,:]
      smallGTimg = GT[j,:,:]
      smallRestoredimg = Restored[j,:,:]
        
      smallLowMSE = mse(smallLowimg, smallLowimg)
      smallLowssim = ssim(smallLowimg, smallLowimg, data_range=smallLowimg.max() - smallLowimg.min())
      
      LowMSE.append(smallLowMSE)
      LowSSIM.append(smallLowssim)
    
      smallGTMSE = mse(smallLowimg, smallGTimg)
      smallGTssim = ssim(smallLowimg, smallGTimg, data_range=smallGTimg.max() - smallGTimg.min())
      
      GTMSE.append(smallGTMSE)
      GTSSIM.append(smallGTssim)
        
      smallRestoredMSE = mse(smallLowimg, smallRestoredimg)
      smallRestoredssim = ssim(smallLowimg, smallRestoredimg, data_range=smallRestoredimg.max() - smallRestoredimg.min())
      
      RestoredMSE.append(smallRestoredMSE)
      RestoredSSIM.append(smallRestoredssim)  
      
```

```python
from bokeh.models import Label
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.io import export_png, output_notebook
hist, edges = np.histogram(np.abs(LowMSE), density=False, bins=10)
p = figure(title="Max Count MeanSquared Error = " + str('%.1f'%edges[np.argmax(hist)]), y_axis_label='Counts',x_axis_label='MSE')
listedges = edges.tolist()
listhist = hist.tolist()

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
    fill_color="#036564", line_color="#033649")
for  i in range(0,len(listhist)):
 if listhist[i] > 0:
  mytext = Label(x=listedges[i], y=listhist[i], text=str('%.2f'%listedges[i]))
  p.add_layout(mytext)
output_notebook()    
show(p)


hist, edges = np.histogram(np.abs(LowSSIM), density=False, bins=10)
p = figure(title="Max Count SSIM = " + str('%.1f'%edges[np.argmax(hist)]), y_axis_label='Counts',x_axis_label='SSIM')
listedges = edges.tolist()
listhist = hist.tolist()

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
    fill_color="#036564", line_color="#033649")
for  i in range(0,len(listhist)):
 if listhist[i] > 0:
  mytext = Label(x=listedges[i], y=listhist[i], text=str('%.2f'%listedges[i]))
  p.add_layout(mytext)
output_notebook()    
show(p)



```
